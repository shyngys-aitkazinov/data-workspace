```python
def k_shagram(s, k):
    ans = []
    for c in s:
        # print(alphabet.index(c) + k)
        new_c = alphabet[(alphabet.r(c) + k) % 26]
        ans.append(new_c)
    ret_s = ''.join(ans)
    ret_w = sum([alphabet.index(c) for c in ret_s])
    return ''.join(ans), ret_w
```

```python
s = 'unequivocally'
for k in range(0, 27, 2):
    print(k, k_shagram(s, k))
```
