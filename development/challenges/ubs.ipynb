```python
import regex as re

def extract_from_string(text):
    # Regular expression pattern
    pattern_canton = r"\s([A-Z]{2})$"
    pattern_zip = r"(\d{4})( |,|$)"

    modified_text, matched_letters_canton, matched_letters_zip = text, None, None

    match_canton = re.search(pattern_canton, text)
    
    if match_canton:
        # Extract the two letters
        matched_letters_canton = match_canton.group(1)

        # Remove the match
        modified_text = re.sub(pattern_canton, "", text).rstrip(',')
    
    match_zip = re.search(pattern_zip, modified_text)

    if match_zip:
        # Extract the four digits
        matched_letters_zip = match_zip.group(1)

        # Remove the match
        modified_text = re.sub(pattern_zip, " ", modified_text)

    city = None
    if ',' in modified_text:
        before_comma, after_comma = modified_text.rsplit(',', 1)  # rsplit splits from the right
        modified_text = before_comma.strip()  # Remove extra spaces around the remaining string
        city = after_comma.strip()

    house_number = None
    pattern_number = r"\b\d+[a-zA-Z]?\d*(\.\d+)?\b"
    match_number = re.search(pattern_number, modified_text)

    if match_number:
        # Extract house number
        house_number = match_number.group(0)

        # Remove the match
        modified_text = re.sub(pattern_number, " ", modified_text)

    return modified_text.replace(".", "").strip(), matched_letters_canton, matched_letters_zip, city, house_number

print(extract_from_string("Zielwg. 256, 4497 Runenverg, BL"))


df1['street']  = df1['decoded_address'].apply(lambda x: extract_from_string(x)[0])
df1['canton'] = df1['decoded_address'].apply(lambda x: extract_from_string(x)[1])
df1['zip'] = df1['decoded_address'].apply(lambda x: extract_from_string(x)[2])
df1['city'] = df1['decoded_address'].apply(lambda x: extract_from_string(x)[3])
df1['house_number'] = df1['decoded_address'].apply(lambda x: extract_from_string(x)[4])



print(df1.head(20))
```

```python
from rapidfuzz import fuzz, process
from tqdm import tqdm


threshold_city_canton = 85
threshold_street = 91
    
def get_best_match(row: pd.Series):
    # First-level filter: City and Canton similarity
    # address_combined = row['city'] + ' ' + row['canton']
    # print(row)
    # Filter potential matches
    # potential_matches = df2.loc[
    #     df2['city_canton'].apply(lambda x: fuzz.token_set_ratio(x, address_combined)) >= threshold_city_canton
    # ]
    potential_matches = df2_subtables[row['zip']] if row['zip'] in df2_subtables else None
    
    if potential_matches is None:
        return None
    
    # If no potential matches, return None
    if potential_matches.empty:
        return None
    
    # print(potential_matches)
    # Second-level filter: Match by street using process.extract
    # concat_matches = potential_matches['street'].apply() + ' ' + potential_matches['house_number']
    # Ensure the 'street' and 'house_number' columns are converted to strings
    potential_matches['street'] = potential_matches['street'].apply(lambda x: str(x) if x is not None else "")
    potential_matches['house_number'] = potential_matches['house_number'].apply(lambda x: str(x) if x is not None else "")

    # Concatenate 'street' and 'house_number' with a space
    concat_matches = potential_matches['street'] + ' ' + potential_matches['house_number']

    ad = ""
    if row['street'] != None:
        ad += str(row['street'])
        
    if row['house_number'] != None:
        ad += " " + str(row['house_number'])
    
    # print(concat_matches)
    matches = process.extract(
        ad, 
        concat_matches.tolist(), 
        scorer=fuzz.token_set_ratio, 
        score_cutoff=threshold_street, 
        limit=5
    )
    
    # print(matches)
    if matches:
        for match in matches:
            idx = concat_matches.index[match[2]]
            if df2.loc[idx]['house_number'] == row['house_number']:
                return df2.loc[idx]['return_address']
                
        return None
    
    
    return None
    # Return the best match if found

tqdm.pandas()
df1['matches'] = df1.progress_apply(lambda row: get_best_match(row), axis=1)



# 105
```
