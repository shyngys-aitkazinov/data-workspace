```python
import pandas as pd
import os
import re

# Path to your folder
folder_path = 'outputs'

# Initialize lists to hold DataFrames
dfs_it = []
dfs_es = []

# Regex to match only non-duplicate files like 'IT_123.csv' or 'ES_456.csv'
valid_file_pattern = re.compile(r'^(IT|ES)_(\d+)\.csv$')

# Iterate over files in the folder
for filename in os.listdir(folder_path):
    if valid_file_pattern.match(filename):
        full_path = os.path.join(folder_path, filename)

        # Read CSV with datetime index
        df = pd.read_csv(full_path, index_col=0, parse_dates=True)

        # Append to the appropriate list
        if filename.startswith('IT_'):
            dfs_it.append(df)
        else:  # starts with ES_
            dfs_es.append(df)

# Combine all dataframes by column
df_it = pd.concat(dfs_it, axis=1)
df_es = pd.concat(dfs_es, axis=1)
```
